```
getGEOSuppFiles("GSE72251")
untar("GSE72251/GSE72251_RAW.tar", exdir = "GSE72251/idat")
head(list.files("GSE72251/idat", pattern = "idat"))
```

```
idatFiles <- list.files("GSE72251/idat", pattern = "idat.gz$", full = TRUE)
sapply(idatFiles, gunzip, overwrite = TRUE)
```

```
rgSet <- read.metharray.exp("GSE72251/idat")
gc()
grSet <- preprocessQuantile(rgSet)
```

```
beta <- getBeta(grSet)
write.csv(file="GSE72251_bvalues.csv", x=beta)
```
